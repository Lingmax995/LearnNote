## 张量 的基本性质

```python
A = torch.arange(20, dtype=torch.float32).reshape(5, 4)
B = A.clone()  # 通过分配新内存，将A的一个副本分配给B
A, A + B
```

```python
#两个矩阵的按元素乘法称为Hadamard积（Hadamard product）
A * B
```

```python
# 将张量乘以或加上一个标量不会改变张量的形状，
# 其中张量的每个元素都将与标量相加或相乘。
a = 2
X = torch.arange(24).reshape(2, 3, 4)
a + X, (a * X).shape
```

## 降维

```python
#求和
x = torch.arange(4, dtype=torch.float32)
x, x.sum()
```

```python
# 我们可以表示任意形状张量的元素和
print(A)
A.sum(),A.sum(axis=0), A.sum(axis=1),A.shape
```

默认情况下，调用求和函数会沿所有的轴降低张量的维度，使它变为一个标量。 我们还可以指定张量沿哪一个轴来通过求和降低维度。 

```python
# 以矩阵为例，为了通过求和所有行的元素来降维（轴0），
# 可以在调用函数时指定axis=0。
# 由于输入矩阵沿0轴降维以生成输出向量，因此输入轴0的维数在输出形状中消失。
A_sum_axis0 = A.sum(axis=0)
A_sum_axis0, A_sum_axis0.shape
```

```python
#指定axis=1将通过汇总所有列的元素降维（轴1）。
# 因此，输入轴1的维数在输出形状中消失。
A_sum_axis1 = A.sum(axis=1)
A_sum_axis1, A_sum_axis1.shape
```

```python
#沿着行和列对矩阵求和，等价于对矩阵的所有元素进行求和。
A.sum(axis=[0, 1])  # 结果和A.sum()相同
```

```python
#可以调用函数来计算任意形状张量的平均值。
A.mean(), A.sum() / A.numel()
```

```python
#计算平均值的函数也可以沿指定轴降低张量的维度。
A.mean(axis=0), A.sum(axis=0) / A.shape[0]
```

## 非降维求和

```python
A
```

```python
#调用函数来计算总和或均值时保持轴数
sum_A = A.sum(axis=1, keepdims=True)
sum_A
```

```python
#由于sum_A在对每行进行求和后仍保持两个轴，我们可以通过广播将A除以sum_A。
A / sum_A
```

```python
#沿某个轴计算A元素的累积总和， 比如axis=0（按行计算），可以调用cumsum函数。 
# 此函数不会沿任何轴降低输入张量的维度。
A.cumsum(axis=0)
```

两个矩阵的按元素乘法被称为他们的Hadamard积。它与矩阵乘法不同。
